import important library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/Student_Attendance.csv')
```

Calculate Attendance %

```python
df['AttendancePercent'] = round((df['ClassesAttended'] / df['TotalClasses']) * 100, 2)
```

Flag Attendance Category

```python
def categorize(p):
    if p >= 75:
        return 'High (>=75%)'
    elif p >= 50:
        return 'Medium (50-74%)'
    else:
        return 'Low (<50%)'

df['Category'] = df['AttendancePercent'].apply(categorize)

df['Flag'] = df['AttendancePercent']
```

Display Final Table

```python
print("\nðŸ“‹ Final Attendance Table:\n")
print(df[['StudentID', 'Name', 'TotalClasses', 'ClassesAttended', 'AttendancePercent', 'Category', 'Flag']])
```

 Bar Chart: Attendance % per student

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Name', y='AttendancePercent', data=df, palette='coolwarm')
plt.title('Attendance Percentage per Student')
plt.xticks(rotation=45)
plt.ylabel('Attendance %')
plt.xlabel('Student')
plt.tight_layout()
plt.show()
```

 Pie Chart: Attendance Category Split

```python
plt.figure(figsize=(6, 6))
df['Category'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['#4CAF50', '#FFC107', '#F44336'])
plt.title('Attendance Category Distribution')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
